```python
import tkinter as tk

# Create the main window
window = tk.Tk()
window.title("GUI")

# Thought label
thought = tk.Label(window, text="An ounce of prevention is worth a pound of cure", bg="light blue", fg="black", font=("Arial", 15), width=120)
thought.pack()

# Main heading
heading = tk.Label(window, text="SYMPTOM SENSE", fg="black", font=("Courier", 50, "bold"), width=15)
heading.pack(padx=0, pady=10)

# Function to show placeholder
def show_placeholder(event=None):
    if entry.get() == "":
        entry.insert(0, placeholder_text)
        entry.config(fg='grey')

# Function to hide placeholder
def hide_placeholder(event):
    if entry.get() == placeholder_text:
        entry.delete(0, tk.END)
        entry.config(fg='black')

# Placeholder text
placeholder_text = "Enter Your Symptoms"

# Entry box
entry = tk.Entry(window, width=90)
entry.pack(pady=20, ipady=10.5)

# Insert placeholder text initially
show_placeholder()

# Bind focus in and out events to hide/show placeholder

entry.bind("<FocusOut>", show_placeholder)

# Run the application
window.mainloop()
```

```python
# Install necessary package
!pip install requests

# Import necessary modules
import tkinter as tk
import requests

# Create the main window
window = tk.Tk()
window.title("Symptom Sense")

# Add a thought-provoking quote
thought = tk.Label(window, text="An ounce of prevention is worth a pound of cure", bg="#5DC6AA", fg="black", font=("Arial", 15), width=80, wraplength=600)
thought.pack(padx=20, pady=20)

# Add a heading
heading = tk.Label(window, text="SYMPTOM SENSE", fg="black", font=("Courier", 50, "bold"))
heading.pack(padx=20, pady=20)

# Add a label prompting the user to enter symptoms
say = tk.Label(window, text="Enter your symptoms", font=("Arial", 14))
say.pack(padx=20, pady=10)

# Add a text box for user input
entry = tk.Text(window, width=80, height=5)
entry.pack(padx=20, pady=10)

# Define the function to handle the button click
def clicked(): 
    user_input = entry.get("1.0", "end-1c").strip()
    if not user_input:
        result.config(text="Please enter your symptoms.", fg="red")
        return

    symptoms = {'symptoms': user_input}
    
    try:
        response = requests.post('http://127.0.0.1:5000/predict', data=symptoms)
        response.raise_for_status()
        
        data = response.json()
        if 'disease' in data:
            result.config(text=f"You are diagnosed with: {data['disease']}", fg="black")
        else:
            result.config(text="Diagnosis not found. Please try again.", fg="red")
    except requests.exceptions.RequestException as e:
        result.config(text=f"Network error: {e}", fg="red")
    except ValueError:
        result.config(text="Invalid response from server.", fg="red")

# Add a submit button
submit = tk.Button(window, text="SUBMIT", font=("Helvetica", 10), bg="light blue", command=clicked)
submit.pack(padx=20, pady=10)

# Add a label to display the result
result = tk.Label(window, text="", bg="yellow", fg="black", width=40, height=10, wraplength=300)
result.pack(padx=40, pady=40)

# Run the Tkinter event loop
window.mainloop()
```
